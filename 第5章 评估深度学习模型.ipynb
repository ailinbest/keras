```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
seed = 7
np.random.seed()

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
```

```python
dataset = np.loadtxt('./pima-indians-diabetes.data.csv',delimiter=',')
```

```python
pd.read_csv('./pima-indians-diabetes.data.csv',header=None).head()
```

```python
X = dataset[:,:8]
y = dataset[:,8]
```

```python
# 创建模型
model = Sequential()
model.add(Dense(units=12,input_dim=8,activation='relu'))
model.add(Dense(units=8,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
# 模型编译
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### 自动评估

```python
# 模型训练，并用0.2的验证集（其实是测试集的功能）
model.fit(x=X,y=y,epochs=150,batch_size=10,validation_split=0.2)
```

- 与没有添加参相比，数增加了val_loss和val_acc两个指标

## 指定验证数据集

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=seed)
```

```python
# 创建模型
model_data = Sequential()
model_data.add(Dense(units=12,input_dim=8,activation='relu'))
model_data.add(Dense(units=8,activation='relu'))
model_data.add(Dense(units=1,activation='sigmoid'))
```

```python
model_data.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

### 手动评估
- 用validation_data参数

```python
# validation_data参数指定验证集
model_data.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=150,batch_size=10)
```

### 手动评估
- 用k折交叉验证

```python
kfold = StratifiedKFold(n_splits=10,shuffle=True,random_state=seed)
```

```python
vcscores = []
# kfold.split()返回的是数据的索引
for train,validation in kfold.split(X=X,y=y):
    model = Sequential()
    model.add(Dense(units=12,input_dim=8,activation='relu'))
    model.add(Dense(units=8,activation='relu'))
    model.add(Dense(units=1,activation='sigmoid'))
    
    #编译模型
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    
    #训练模型
    model.fit(X[train],y[train],epochs=100,batch_size=10,verbose=0)
    
    # 评估模型
    scores = model.evaluate(X[validation],y[validation],verbose=0)
    
    # 输出评估结果
    print('%s: %.2f%%' % (model.metrics_names[1],scores[1]*100))
    vcscores.append(scores[1]*100)

# 输出accuracy的均值和标准差
print('accuracy的均值为 %.2f%%，标准差为 +/- %.2f%%' % (np.mean(vcscores),np.std(vcscores)) )
```
